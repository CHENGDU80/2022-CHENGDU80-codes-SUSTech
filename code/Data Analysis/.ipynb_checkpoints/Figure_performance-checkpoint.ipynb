```python
models = ['lr', 'svm', 'mlp','lgbm', 'xgb']
accuracy = [0.941, 0.958, 0.963, 0.984,0.967]
auc = [0.547, 0.567, 0.539, 0.612, 0.543]
f1 = [0.493, 0.544, 0.485, 0.516, 0.524]
cols = ['Accuracy', 'AUC', 'F1-Score']
```

```python
import matplotlib.pyplot as plt
```

```python
data = [accuracy[0], auc[0], f1[0]]
# plt.figure(10, 15)
plt.bar(range(len(data)), data)
plt.xticks(range(len(cols)), cols, rotation=-45, fontsize=14)
plt.title("Linear Regression")
plt.show()
```
