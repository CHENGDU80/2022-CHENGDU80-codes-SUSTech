```python
def woe_i(good_i, bad_i, good_t, bad_t):
    return math.log(math.e, bad_i / bad_t) - math.log(math.e, good_i / good_t)

def iv_i(good_i, bad_i, good_t, bad_t, woe_i):
    return woe_i * (bad_i / bad_t - good_i / good_t)
```

```python
def woe0iv(data, target_col, num_bins, label_col, bad_t, good_t):
    s_woe = []
    s_iv = []
    df = data[[target_col, 'DEFAULT_LABEL']]
    aim_df = df.sort_values(by=target_col, inplace=False)
    
#     vc = aim_df[label_col].value_counts()
    # Nan out
    cnt = 0
    step = int(len(train) / num_bins)
    for i in range(num_bins):
        if i == num_bins - 1:
            bad_i = int(aim_df['DEFAULT_LABEL'][cnt:].value_counts()[1])
            good_i = int(aim_df['DEFAULT_LABEL'][cnt:].value_counts()[0])
        else:
            bad_i = int(aim_df['DEFAULT_LABEL'][cnt:cnt+step].value_counts()[1])
            good_i = int(aim_df['DEFAULT_LABEL'][cnt:cnt+step].value_counts()[0])
            cnt += step
        woe_i_res = woe_i(good_i, bad_i, good_t, bad_t)
        s_woe.append(woe_i_res)
        s_iv.append(iv_i(good_i, bad_i, good_t, bad_t, woe_i_res))
    s_iv.append(sum(s_iv))
    return s_woe, s_iv
#     print(len(vc))
```

```python
def infogain(data, target_col, init_E):
    IG = init_E
    df = data[[target_col, 'DEFAULT_LABEL']]
    aim_df = df.sort_values(by=target_col, inplace=False)
    vc = aim_df[target_col].value_counts()
    index = 0
    for i in range(len(vc)):
        aim_i = aim_df[index:index+vc[i]]
        if len(aim_i['DEFAULT_LABEL'].value_counts()) == 1:
            index += vc[i]
            continue
        else:
            bad = int(aim_i['DEFAULT_LABEL'].value_counts()[1])
            good = int(aim_i['DEFAULT_LABEL'].value_counts()[0])
        E_i = - bad / vc[i] * math.log(2, bad / vc[i]) - good / vc[i] * math.log(2, good / vc[i])
        IG -= vc[i] / len(train) * E_i
        index += vc[i]
    return IG
```

```python
def bin_distance(x,y,n=10): # x为待分箱的变量，y为target变量.n为分箱数量
    total = y.count()  # 计算总样本数
    bad = y.sum()      # 计算坏样本数
    good = y.count()-y.sum()  # 计算好样本数
    d1 = pd.DataFrame({'x':x,'y':y,'bucket':pd.cut(x,n)}) #利用pd.cut实现等距分箱
    d2 = d1.groupby('bucket',as_index=True)  # 按照分箱结果进行分组聚合
    d3 = pd.DataFrame(d2.x.min(),columns=['min_bin']) 
    d3['min_bin'] = d2.x.min()  # 箱体的左边界
    d3['max_bin'] = d2.x.max()  # 箱体的右边界
    d3['bad'] = d2.y.sum()  # 每个箱体中坏样本的数量
    d3['total'] = d2.y.count() # 每个箱体的总样本数
    d3['bad_rate'] = d3['bad']/d3['total']  # 每个箱体中坏样本所占总样本数的比例
    d3['badattr'] = d3['bad']/bad   # 每个箱体中坏样本所占坏样本总数的比例
    d3['goodattr'] = (d3['total'] - d3['bad'])/good  # 每个箱体中好样本所占好样本总数的比例
    d3['woe'] = np.log(d3['goodattr']/d3['badattr'])  # 计算每个箱体的woe值
    iv = ((d3['goodattr']-d3['badattr'])*d3['woe']).sum()  # 计算变量的iv值
    d4 = (d3.sort_values(by='min_bin')).reset_index(drop=True) # 对箱体从大到小进行排序
    print('分箱结果：')
    print(d4)
    print('IV值为：')
    print(iv)
    cut = []
    cut.append(float('-inf'))
    for i in d4.min_bin:
        cut.append(i)
    cut.append(float('inf'))
    woe = list(d4['woe'].round(3))
    return d4,iv,cut,woe
```
