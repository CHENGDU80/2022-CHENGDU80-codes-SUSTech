```python
import numpy as np
import pandas as pd
import math
```

```python
X_train = pd.read_csv("./data/cleaned/X_train.csv")
X_train = X_train.drop(X_train.columns[[0]], axis = 1)
y_train = pd.read_csv("./data/cleaned/y_train.csv")
y_train = y_train.drop(y_train.columns[[0]], axis = 1)
merge_train = X_train.join(y_train)
```

```python
normal_train = merge_train.loc[merge_train['DEFAULT_LABEL']==0]
default_train = merge_train.loc[merge_train['DEFAULT_LABEL']==1]
normal_train=normal_train.describe().T
normal_train.drop(normal_train.columns[[0,3,4,5,6,7]], axis = 1,inplace=True)
normal_train.index.name = 'Variables'
normal_train=normal_train.reset_index()
normal_train=normal_train.rename(columns={'mean':'normal_mean','std':'normal_std'})
default_train=default_train.describe().T
default_train.drop(default_train.columns[[0,3,4,5,6,7]], axis = 1,inplace=True)
default_train.index.name = 'Variables'
default_train=default_train.reset_index()
default_train=default_train.rename(columns={'mean':'default_mean','std':'default_std'})
```

```python
table1 = X_train.describe().T
table1.drop(table1.columns[[0]], axis = 1)
table1.index.name = 'Variables'
table1.reset_index()
```

```python
table2 = pd.read_csv("./data/datapreview.csv")
table2
```

```python
merge_table = pd.merge(table1,table2,how='left',left_on='Variables',right_on='Variables')
merge_table = pd.merge(merge_table,normal_train,how='left',left_on='Variables',right_on='Variables')
merge_table = pd.merge(merge_table,default_train,how='left',left_on='Variables',right_on='Variables')
merge_table.drop(merge_table.columns[[1]], axis = 1,inplace=True)
merge_table
```

```python
merge_table.to_csv("./data/datapreview2.csv",index=False)
```
